```
import numpy as np

def echelon(matrice):
  
    pivot_position = 0
    
    for row_number in range(matrice.shape[0]):
      
        temp = row_number
        
        while matrice[temp, pivot_position] == 0:
            temp += 1
            if matrice.shape[0] == temp:
                temp = row_number
                pivot_position += 1
                if matrice.shape[0] == pivot_position:
                    return False
                  
        matrice[[temp, row_number]] = matrice[[row_number, temp]]
        
        if matrice[row_number, pivot_position] != 0:
            matrice[row_number] = matrice[row_number] / matrice[row_number, pivot_position]
            
        for i in range(matrice.shape[0]):
            if i != row_number:
                matrice[i] = matrice[i] - (matrice[row_number] * matrice[i, pivot_position])
                
        pivot_position += 1
        
        print("step " + str(row_number + 1) + ": ")
        print(matrice)
        
    return True

  
n = int(input("enter n: "))
A = np.empty((0, n))

for i in range(n):
  A = np.vstack((A, list(input("enter row " + str(i + 1) + " of A: ").split())))

b = np.empty((0, 1))
for i in range(n):
  b = np.vstack((b, list(input("enter row " + str(i + 1) + " of b: ").split())))


augmented_matrice = np.column_stack((A, b))
augmented_matrice = augmented_matrice.astype(np.float)

print("This is A|b:")
print(augmented_matrice)

if echelon(np.column_stack((A, b)).astype(np.float)):
    print("This is final answer of A|b:")
    print(np.linalg.solve(A.astype(np.float), b.astype(np.float)))

print("This is upper triangular of A|b:")
upper_triangular = np.triu(augmented_matrice)
print(upper_triangular)
```
